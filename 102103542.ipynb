```python
from pytube import YouTube
from moviepy.editor import VideoFileClip
from pydub import AudioSegment

video_urls = [
    "https://www.youtube.com/watch?v=e-ORhEE9VVg",
    "https://www.youtube.com/watch?v=8xg3vE8Ie_E",
    "https://www.youtube.com/watch?v=VuNIsY6JdUw",
    "https://www.youtube.com/watch?v=w1oM3kQpXRo",
    "https://www.youtube.com/watch?v=j7_9x26CUyE",
    "https://www.youtube.com/watch?v=-BjZmE2gtdo",
    "https://www.youtube.com/watch?v=sqquaClyS0A",
    "https://www.youtube.com/watch?v=bWjf_xIHvQ0",
    "https://www.youtube.com/watch?v=Q6vxs_d7eCo",
    "https://www.youtube.com/watch?v=D6Jd8-eFX8k",
    "https://www.youtube.com/watch?v=Q6vxs_d7eCo"
    
]

for url in video_urls:
    try:
        yt = YouTube(url)
        
        stream = yt.streams.get_highest_resolution()
        video_path = stream.download()
        
        print(f"Downloaded: {yt.title}")
        
        mp3_path = video_path[:-4] + ".mp3" 
        video = VideoFileClip(video_path)
        video.audio.write_audiofile(mp3_path)
        
        print(f"Converted to MP3: {mp3_path}")
        
        audio = AudioSegment.from_file(mp3_path)
        
        start_time = 10000  
        end_time = len(audio)  
        
        trimmed_audio = audio[start_time:end_time]
        
        trimmed_audio.export(mp3_path, format="mp3")
        
        print("Trimmed audio saved successfully")
        
    except Exception as e:
        print(f"Error downloading/converting {url}: {e}")
```

```python
from pydub import AudioSegment

def merge_audio_files(file_paths, output_path):
    
    merged_audio = AudioSegment.empty()
    
    
    for file_path in file_paths:

        audio = AudioSegment.from_file(file_path)
        
        merged_audio += audio
    
    merged_audio.export(output_path, format="mp3")  

if __name__ == "__main__":
    file_paths = [
        "/kaggle/working/Cruel Summer - Taylor Swift Lyrics.mp3",
        "/kaggle/working/Blank Space.mp3",
        "/kaggle/working/One Way Or Another (Teenage Kicks).mp3",
        "/kaggle/working/See You Again.mp3",
        "/kaggle/working/Love Story.mp3",
        "/kaggle/working/Story of My Life (Official 4K Video).mp3",
        "/kaggle/working/Story of My Life (Official 4K Video).mp3",
        "/kaggle/working/Summertime Sadness (Lyrics) - Lana Del Rey.mp3",
        "/kaggle/working/Perfect (Official Video).mp3",
        "/kaggle/working/Lover (Official Music Video).mp3",
        
        
    ]
    
    output_path = "/kaggle/working/vmerged_audio.mp3"
    
    merge_audio_files(file_paths, output_path)
```
